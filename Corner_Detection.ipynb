# Harris Corner Detection

```python
gray = np.float32(gray_chess)
```

```python
dst = cv2.cornerHarris(src = gray, blockSize =10, ksize =5, k = 0.04)
```

```python
dst = cv2.dilate(dst, None)
```

```python
chess[dst > 0.01*dst.max()] = [255,0,0]
```

```python
plt.imshow(chess)
```
