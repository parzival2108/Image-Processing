```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
plt.figure(figsize=(10,10))
img1 = cv2.imread("leth.jpeg")
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
plt.imshow(img1)
```

```python
def load_img():
    
    img1 = cv2.imread("leth.jpeg").astype(np.float32)/255
    img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
    return img1
    
```

```python
def display_img(img1):
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(111)
    ax.imshow(img1)
```

```python
i = load_img()
display_img(i)
```
