### Thresholding is segmenting image into different parts, it will convert image to consist of only two values, white or black

```python
plt.figure(figsize=(10,10))
img1 = cv2.imread("leth.jpeg")
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
plt.imshow(img1)
```

```python
img2 = cv2.imread("watermark.png")
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
plt.imshow(img2)
```
