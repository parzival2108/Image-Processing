```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
plt.figure(figsize=(10,10))
img1 = cv2.imread("leth.jpeg")
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
plt.imshow(img1)
```

```python
def load_img():
    
    img1 = cv2.imread("leth.jpeg").astype(np.float32)/255
    img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
    return img1
    
    
```

```python
#load_img()
```

```python
def display_img(img1):
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(111)
    ax.imshow(img1)
```

```python
i = load_img()
display_img(i)
```

# Gamma Correction

```python
gamma = 2
```

```python
result = np.power(i,gamma)
display_img(result)
```

# Low pass filter with 2D convolution

```python
img2 = load_img()
font = cv2.FONT_HERSHEY_COMPLEX
cv2.putText(img2, text = "Lethi", org = (150,350), fontFace = font, fontScale = 15, color = (255,0,0), thickness =4)
```

```python
display_img(img2)
```

```python
kernel = np.ones(shape = (5,5), dtype = np.float32)/50
```

```python
kernel
```

```python
dst = cv2.filter2D(img2, -1, kernel)
display_img(dst)
```

# Blurring

```python
img2 = load_img()
font = cv2.FONT_HERSHEY_COMPLEX
cv2.putText(img2, text = "Lethi", org = (150,350), fontFace = font, fontScale = 15, color = (255,0,0), thickness =4)
print("reset")
```

```python
blurred = cv2.blur(img2, ksize = (20,20))
display_img(blurred)
```

# Gaussian Blurr

```python
blurred_image = cv2.GaussianBlur(img2, (19,19), 10)
```
